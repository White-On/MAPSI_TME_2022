# Simple Plot

```python
def f(x : np.array) -> np.array:
    return np.sin(x) + np.cos(x) * x

xValues = np.linspace(0, 100, 1000)
yValues = f(xValues)
plt.plot(xValues, yValues)
plt.show()
```

# Multiple Plots

```python
de1 = np.random.choice(range(1,7), size=10000)
de2 = np.random.choice(range(1,7), size=10000)

gridsize = (3, 2)
fig = plt.figure(figsize=(12,8))
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=2,rowspan=2)
ax2 = plt.subplot2grid(gridsize, (2, 0))
ax3 = plt.subplot2grid(gridsize, (2, 1))
ax2.hist(de1, bins=6,edgecolor='black')
ax3.hist(de2, bins=6,edgecolor='black')
ax1.hist(de1+de2, bins=11, edgecolor='black')

plt.show()
```

# Animated Simple plot 

```python
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'ro')

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)
plt.show()
```

# Stack Plot

Permet de crée une supperposition de plusieurs graphiques pour étudier la somme des valeurs.

```python
rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))
yrs = 1950 + rng

fig, ax = plt.subplots(figsize=(10, 6))
ax.stackplot(yrs, rng + rnd, labels=['Eastasia', 'Eurasia', 'Oceania'])
ax.set_title('Combined debt growth over time')
ax.legend(loc='upper left')
ax.set_ylabel('Total debt')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])
fig.tight_layout()
```

# More Plot

```python
textstr1 = '\n'.join((
    r'$\mu=%.2f$' % (np.mean(res), ),
    r'$\sigma=%.2f$' % (np.std(res), ),
    r'$\varphi=%.2f$' % (np.var(res), )))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 6))
ax = axes[0]
ax.grid(True,lw=0.5,ls='--')
ax.hist(res, bins=30, density=True, histtype='step', alpha=0.5, label="a")
ax.hist(res2, bins=30, density=True, histtype='step', alpha=0.5, label="b")
ax.legend(loc='upper right', fancybox= False, edgecolor='white', fontsize=14)
ax.text(0.05,0.95,textstr1,transform=ax.transAxes,fontsize=14,verticalalignment='top',
            bbox = dict(facecolor='black', edgecolor ='white')) #transform permet de définir la position du texte
ax.set_ylabel('Density')
fig.text(0.5, 0.04, '$\Delta E$ [Joules]', ha='center', va='center', fontsize=20)
plt.show()
```

# 2D Plot

```python
_ = np.linspace(-1, 1 , 100)
x, y = np.meshgrid(_, _)
z = x**2 + y*x
x.shape, y.shape, z.shape
```

## Filled-in contour plot

```python
plt.contourf(x, y, z, 20, cmap='RdGy', levels = 50) # level permet de définir le nombre de précision et de couleurs
plt.colorbar(label=('$z$'))
plt.show()
```

## Contour plot

```python
cs = plt.contour(x, y, z, 20)
plt.clabel(cs, inline=True, fontsize=10)
```

# 3D Plot

```python
fig , ax = plt.subplots(figsize=(10, 6), subplot_kw=dict(projection='3d'))
ax.plot_surface(x, y, z, cmap='viridis')
```
